> 정규성검증

```python
def normality_test(df, tool='Shapiro'):

    from scipy.stats import shapiro, anderson, kstest,jarque_bera, normaltest,anderson_ksamp

    result_dict = {'col':[], 'p_value':[], 't_stat':[]}

    if tool == 'Shapiro':
        for i in df.columns:
            statistic, p_value = shapiro(df[i])
            result_dict['col'].append(i)
            result_dict['p_value'].append(p_value)
            result_dict['t_stat'].append(statistic)

            test = pd.DataFrame(result_dict)
    
    elif tool == 'k_s':
        for i in df.columns:
            statistic, p_value = kstest(df[i], 'norm')
            result_dict['col'].append(i)
            result_dict['p_value'].append(p_value)
            result_dict['t_stat'].append(statistic)

            test = pd.DataFrame(result_dict)

    elif tool == 'jarque_bera':
        for i in df.columns:
            statistic, p_value = jarque_bera(df[i])
            result_dict['col'].append(i)
            result_dict['p_value'].append(p_value)
            result_dict['t_stat'].append(statistic)

            test = pd.DataFrame(result_dict)


    elif tool == 'anderson':

        result_dict = {'col' : [], 'critical_values' : [], 't_stat' : []}

        for i in df.columns:
            result = anderson(df[i])
            result_dict['col'].append(i)
            result_dict['critical_values'].append(result.critical_values[2])
            result_dict['t_stat'].append(result.statistic)

            test = pd.DataFrame(result_dict)

    return test
```

> 등분산 검정

```python
from scipy.stats import bartlett
```

```python
# 등분산 검정

# 등분산 검정 결과를 저장할 데이터프레임 생성
result_df = pd.DataFrame(columns=['Variable', 'Group_0_p-value', 'Group_1_p-value'])

# 각 독립변수에 대해 등분산 검정 수행
for variable in df.columns[:-1]:
    group_0_data = df.loc[df['t-1감사의견코드'] == 0, variable]
    group_1_data = df.loc[df['t-1감사의견코드'] == 1, variable]
    
    t_value, p_value = bartlett(group_0_data, group_1_data)
    result_df = result_df.append({'Variable': variable, 'Group_0_p-value': p_value, 'Group_1_p-value': p_value}, ignore_index=True)

# p-value가 0.05보다 큰 값을 필터링하여 출력
filtered_result_df = result_df[result_df['Group_0_p-value'] > 0.05]
filtered_result_df
```

> Welch's t-test

```python
import pandas as pd
from scipy.stats import ttest_ind

# 독립변수, 종속변수 설정
X = df.drop('t-1감사의견코드', axis=1)
y = df['t-1감사의견코드']

# Welch's t-test를 통한 변수 선택
selected_features = []

for feature in X.columns:
    group_0 = X[y == 0][feature]
    group_1 = X[y == 1][feature]

    t_statistic, p_value = ttest_ind(group_0, group_1, equal_var=False)

    if p_value < 0.05:  # 유의수준 0.05로 설정
        selected_features.append({'Variable': feature, 'p-value': p_value})

# 결과를 데이터프레임으로 만들어 출력
result_df_wt = pd.DataFrame(selected_features)

result_df_wt.reset_index()
result_df_wt
```

```python
def vif(data):
    import pandas as pd
    from statsmodels.stats.outliers_influence import variance_inflation_factor

    # VIF 출력을 위한 데이터 프레임 형성
    vif = pd.DataFrame()

    # VIF 값과 각 Feature 이름에 대해 설정
    vif["VIF Factor"] = [variance_inflation_factor(data.values, i) for i in range(len(data.columns))]
    vif["features"] = data.columns 
                    
    # VIF 값이 높은 순으로 정렬
    vif = vif.sort_values(by="VIF Factor", ascending=False)
    vif = vif.reset_index().drop(columns='index')
    
    return vif

vif(df)
```

```python
# Welch's t-test를 통한 변수 선택
selected_features = []

for feature in df.columns:
    group_0 = df[y == 0][feature]
    group_1 = df[y == 1][feature]

    t_statistic, p_value = ttest_ind(group_0, group_1, equal_var=False)

    if p_value < 0.05:  # 유의수준 0.05로 설정
        selected_features.append({'Variable': feature, 'p-value': p_value})

# 결과를 데이터프레임으로 만들어 출력
result_df_wt = pd.DataFrame(selected_features, index=df.columns)

result_df_wt.reset_index()
result_df_wt.index
```
