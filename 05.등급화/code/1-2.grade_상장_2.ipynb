```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('../data/상장raw2.csv', encoding='euc-kr')
df상장 = pd.read_excel('../data/kospi+kodaq_dp(등급).xls')
```

```python
df['회계년도']= df['회계년도'].replace('/','').astype(str).str[:4].astype(int)
df = df[df['회계년도'] == 2022]
```

```python
df['FCF'] = df['영업활동으로 인한 현금흐름(간접법)(*)(IFRS연결)(천원)'] + df['투자활동으로 인한 현금흐름(*)(IFRS연결)(천원)']
df['총자산대비잉여현금흐름'] = df['FCF'] / df['자산(*)(IFRS연결)(천원)']
```

```python
df = df[['회사명', '거래소코드', '총자본회전률(IFRS연결)', '자기자본순이익률(IFRS연결)', '총자본순이익률(IFRS연결)', '부채비율(IFRS연결)', '총자산대비잉여현금흐름', '정상영업이익증가율(IFRS연결)', '자기자본증가율(IFRS연결)', '유동자산회전률(IFRS연결)', '재고자산회전률(IFRS연결)', '총자본투자효율(IFRS연결)', '매출액정상영업이익률(IFRS연결)', '자기자본구성비율(IFRS연결)', '차입금의존도(IFRS연결)']]
```

```python
df['거래소코드'] = df['거래소코드'].astype(str).str.zfill(6)
```

```python
df상장.dropna(axis=0, inplace=True)
```

```python
df상장['Stock'] = df상장['Stock'].astype(int)
```

```python
df상장['Stock'] = df상장['Stock'].astype(str).str.zfill(6)
```

```python
df상장.columns = ['KIS', '거래소코드', '회사명_1', '신용평점']
```

```python
df_total = pd.merge(df,df상장,how='inner',on='거래소코드')
```

```python
df_total['신용평점']=df_total['신용평점'].map({
    'AAA' : 'AA',
    'AA':'AA',
    'A':'A',
    'BBB':'BBB',
    'BB':'BB',
    'B':'B',
    'CCC':'CCC',
    'CC':'CC',
    'C': 'D',
    'D':'D'

})
```

```python
df_total.isna().sum()
```

```python
df_total=df_total.dropna()
```

```python
df_total = df_total.replace([np.inf, -np.inf], np.nan)  # inf 값을 NaN으로 대체
df_total = df_total.dropna()  # NaN이 있는 행 삭제
```

```python
df_total.describe()
```

```python
df_total.drop('회사명_1', axis=1, inplace=True)
```

```python
df_total.columns = ['회사명', '거래소코드', '총자본회전률', '자기자본순이익률', '총자본순이익률',
       '부채비율', '총자산대비잉여현금흐름', '영업이익증가율', '자기자본증가율',
       '유동자산회전률', '재고자산회전률', '총자본투자효율',
       '매출액영업이익률', '자기자본구성비율', '차입금의존도', 'KIS',
       '신용평점']
```

```python
coefficients = {
    '총자본회전률' : 0.88069077,
    '자기자본순이익률' : 0.76604203,
    '총자본순이익률' : 0.47610808,
    '부채비율' : -0.43508945,
    '총자산대비잉여현금흐름' : -0.19130195,
    '영업이익증가율' : 0.03846689,
    '자기자본증가율' : -0.56756804,
    '유동자산회전률' : 0.33787758,
    '재고자산회전률' : 0.25749421,
    '총자본투자효율' : -0.48339344,
    '매출액영업이익률' : -0.27890602,
    '자기자본구성비율' : -0.39173785,
    '차입금의존도' : -0.20683487,
    'constant' : 0.7891004
}


# O-Score 계산
df_total['dp_score'] = coefficients['constant']
df_total['dp_score'] += coefficients['총자본회전률'] * df_total['총자본회전률']
df_total['dp_score'] += coefficients['자기자본순이익률'] * df_total['자기자본순이익률']
df_total['dp_score'] += coefficients['총자본순이익률'] * df_total['총자본순이익률']
df_total['dp_score'] += coefficients['부채비율'] * df_total['부채비율']
df_total['dp_score'] += coefficients['총자산대비잉여현금흐름'] * df_total['총자산대비잉여현금흐름']
df_total['dp_score'] += coefficients['영업이익증가율'] * df_total['영업이익증가율']
df_total['dp_score'] += coefficients['자기자본증가율'] * df_total['자기자본증가율']
df_total['dp_score'] += coefficients['유동자산회전률'] * df_total['유동자산회전률']
df_total['dp_score'] += coefficients['재고자산회전률'] * df_total['재고자산회전률']
df_total['dp_score'] += coefficients['총자본투자효율'] * df_total['총자본투자효율']
df_total['dp_score'] += coefficients['매출액영업이익률'] * df_total['매출액영업이익률']
df_total['dp_score'] += coefficients['자기자본구성비율'] * df_total['자기자본구성비율']
df_total['dp_score'] += coefficients['차입금의존도'] * df_total['차입금의존도']
```

```python
lower_bound = df_total['dp_score'].quantile(0.01)
upper_bound = df_total['dp_score'].quantile(0.99)
df_total_filtered = df_total[(df_total['dp_score'] >= lower_bound) & (df_total['dp_score'] <= upper_bound)]
```

```python
df_total[['dp_score','신용평점']].groupby('신용평점').agg(['mean', lambda x: x.quantile(0.25), lambda x: x.quantile(0.5), lambda x: x.quantile(0.75)])
```
