```python
# 산술평균 함수 만들기
def calculate_derived_value(row):
    
    # 두개의 분기값이 존재하면 산술평균
    if pd.notna(row['시가총액_x']) and pd.notna(row['시가총액_y']):
        return (row['시가총액_x'] + row['시가총액_y']) / 2
    
    # 두개의 분기값 중 하나라도 결측치면 그 값 그대로 대입
    elif pd.notna(row['시가총액_x']):
        return row['시가총액_x']
    elif pd.notna(row['시가총액_y']):
        return row['시가총액_y']
    else:
        return None
```

```python
# 합친 1,2분기 데이터에서 함수 적용하여 산술평균 파생변수 생성
df2013_12['시가총액'] = df2013_12.apply(calculate_derived_value, axis=1)
df2014_12['시가총액'] = df2014_12.apply(calculate_derived_value, axis=1)
df2015_12['시가총액'] = df2015_12.apply(calculate_derived_value, axis=1)
df2016_12['시가총액'] = df2016_12.apply(calculate_derived_value, axis=1)
df2017_12['시가총액'] = df2017_12.apply(calculate_derived_value, axis=1)
df2018_12['시가총액'] = df2018_12.apply(calculate_derived_value, axis=1)
df2019_12['시가총액'] = df2019_12.apply(calculate_derived_value, axis=1)
df2020_12['시가총액'] = df2020_12.apply(calculate_derived_value, axis=1)
df2021_12['시가총액'] = df2021_12.apply(calculate_derived_value, axis=1)
df2022_12['시가총액'] = df2022_12.apply(calculate_derived_value, axis=1)
```
