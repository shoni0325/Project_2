# raw data 만들기
* 순서
    1.  감사의견 없는 연결, 개별, GAAP 합치기
    
        >  연결 -> 개별 -> GAAP 순으로 결측치 대체
    2. 감사의견 있는 연결 파일의 감사의견 컬럼을 새로 만든 통합 데이터프레임에 입력

```python
# 데이터 로드
data_1 = pd.read_csv('../data/외감기업(IFRS연결)_감사의견미포함.csv', encoding='euc-kr')
data_2 = pd.read_csv('../data/외감기업(IFRS개별).csv', encoding='euc-kr')
data_3 = pd.read_csv('../data/외감기업(GAAP).csv', encoding='euc-kr', index_col=0)

data_target = pd.read_csv('../data/외감기업(IFRS연결).csv', encoding='euc-kr')
```

```python
data_3.reset_index(drop=True, inplace=True)
```

```python
# 컬럼명 저장
col_list = data_1.columns.tolist()
```

```python
data_3[['연구비', '경상개발비(요약)(백만원)']]
```

```python
# 연구비, 경상개발비 합치기
data_3['연구비'] = data_3['연구비'] + data_3['경상개발비(요약)(백만원)']
data_3[['연구비', '경상개발비(요약)(백만원)']]
```

```python
data_3.drop('경상개발비(요약)(백만원)', axis=1, inplace=True)
```

```python
# 컬럼명 통일
data_1.columns = [f"X_{i}" for i in range(1, 51)]
```

```python
data_2.columns = [f"X_{i}" for i in range(1, 51)]
```

```python
data_3.columns = [f"X_{i}" for i in range(1, 51)]
```

```python
# IFRS연결, IFRS개별 합치기
raw_data_1 = data_1.fillna(data_2)
```

```python
# IFRS, GAAP 합치기
raw_data_2 = raw_data_1.fillna(data_3)
```

```python
raw_data_2.isnull().sum()
```

```python
raw_data_2.columns = col_list
```
