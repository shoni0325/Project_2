```python
# 필요한 라이브러리 import
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# Create SVC model
model = SVC(kernel='linear', C=1)

# Perform cross-validation
accuracy_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
precision_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='precision')
recall_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='recall')
f1_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='f1')

# Print cross-validation results
print("교차 검증 결과:")
print("평균 정확도:", accuracy_scores.mean())
print("평균 정밀도:", precision_scores.mean())
print("평균 재현율:", recall_scores.mean())
print("평균 F1 스코어:", f1_scores.mean())

# 가장 높은 F1 스코어를 가진 모델을 선택합니다.
best_model_index = f1_scores.argmax()
best_model = SVC(kernel='linear', C=1)
best_model.fit(X_train, y_train)

# Predict and evaluate test data
y_pred = best_model.predict(X_test)

# Calculate the performance evaluation index of the test data
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Print performance indicators of test data
print("테스트 데이터 성능 지표:")
print("정확도:", accuracy)
print("정밀도:", precision)
print("재현율:", recall)
print("F1 스코어:", f1)
print("혼동 행렬:\n", conf_matrix)
```
